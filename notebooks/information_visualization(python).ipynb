Data sampling

```python
import random

population = list(range(1,1000))
sample_size = 10
random_sample = random.sample(population , sample_size)
print("Random Sample : " , random_sample)
```

Starified Sampling

```python
import numpy as np
import pandas as pd
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']*10, #GENERATES 10 COLOUMNS
    'Value': np.random.randn(100)
})
stratified_sample = data.groupby('Category', group_keys=False).apply(lambda x: x.sample(min(len(x), 2)))
print("stratified_sample:\n",stratified_sample)
```

Cluster Sampling

```python
clusters = [list(range(1, 36)),list(range(26, 41)),list(range(51, 72))]
cluster_sample = random.choice(clusters)
print("Cluster Sample:", cluster_sample)
```

Descriptive Statistics

```python
import numpy as np
import pandas as pd
data = pd.DataFrame({
    'Values': np.random.randn(95)
})
summary_stats = data.describe()
print("Summary Statistics:\n",summary_stats)
```

```python
import matplotlib.pyplot as plt
data['Values'].hist(bins=150, edgecolor='black')
plt.title('Histogram of Values')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

Dimensionality Reduction

```python
#principal component analysis
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris=load_iris()
X = iris.data  # Extract the data from the iris dataset
pca=PCA(n_components=2)
reduced_data=pca.fit_transform(X)
plt.scatter(reduced_data[:,0],reduced_data[:,1],c=iris.target)  # Use reduced_data here
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar(label="Species")
plt.show()
```

Parallel Coordinates

```python
#interactive Visualization - Parallel Coordinates and Iterations
#Parallel Coordinates

from pandas.plotting import parallel_coordinates
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/iris-data.csv')
parallel_coordinates(data, 'class', color=['b','g','r'])
plt.title('Parallel Coordinates of Iris Dataset')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```
